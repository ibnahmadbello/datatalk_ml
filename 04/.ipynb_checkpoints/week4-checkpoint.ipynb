```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('course_lead_scoring.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe(include='object')
```

```python
categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
categorical_columns
```

```python
numerical_columns = list(df.dtypes[df.dtypes != 'object'].index)
```

```python
numerical_columns = ['number_of_courses_viewed', 'annual_income', 'interaction_count', 'lead_score']
```

```python
numerical_columns
```

```python
for n in numerical_columns:
    df[n] = df[n].fillna(0)
```

```python
for c in categorical_columns:
    df[c] = df[c].fillna('NA')
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.converted.values
y_val = df_val.converted.values
y_test = df_test.converted.values

del df_train['converted']
del df_val['converted']
del df_test['converted']
```

```python
# Compute AUC for each numerical variable
auc_scores = {}

for feature in numerical_columns:
    # Calculate AUC
    auc = roc_auc_score(y_train, df_train[feature])
    
    # If AUC < 0.5, invert the variable
    if auc < 0.5:
        auc = roc_auc_score(y_train, -df_train[feature])
        auc_scores[feature] = auc
        print(f"{feature}: {auc:.4f} (inverted)")
    else:
        auc_scores[feature] = auc
        print(f"{feature}: {auc:.4f}")

# Find the numerical variable with the highest AUC
best_feature = max(auc_scores, key=auc_scores.get)
highest_auc = auc_scores[best_feature]

print(f"\nThe numerical variable with the highest AUC is '{best_feature}' with AUC = {highest_auc:.4f}")
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical_columns + numerical_columns].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical_columns + numerical_columns].to_dict(orient='records')
X_val = dv.transform(val_dict)

y_pred = model.predict_proba(X_val)[:, 1]
conversion_decision = (y_pred >= 0.5)
(y_val == conversion_decision).mean()
```

```python
# Prepare the data for DictVectorizer
train_dict = df_train.to_dict(orient='records')
val_dict = df_val.to_dict(orient='records')

# Initialize and fit DictVectorizer
dv = DictVectorizer(sparse=False)
X_train_encoded = dv.fit_transform(train_dict)
X_val_encoded = dv.transform(val_dict)

print(f"Number of features after one-hot encoding: {X_train_encoded.shape[1]}")

# Train logistic regression model
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=1)
model.fit(X_train_encoded, y_train)

# Make predictions on validation set
y_val_pred = model.predict_proba(X_val_encoded)[:, 1]

# Calculate AUC on validation set
val_auc = roc_auc_score(y_val, y_val_pred)

print(f"AUC on validation dataset: {val_auc:.3f}")


# Get feature names after one-hot encoding
feature_names = dv.get_feature_names_out()
print(f"\nFirst 20 feature names: {feature_names[:20]}")
print(f"Total number of features: {len(feature_names)}")


# Check if model converged
print(f"Model converged: {model.n_iter_}")
print(f"Number of iterations: {model.n_iter_}")
```

```python
from sklearn.metrics import precision_score, recall_score
# Generate predictions on validation set
y_val_pred_proba = model.predict_proba(X_val_encoded)[:, 1]

# Initialize lists to store precision and recall values
thresholds = np.arange(0.0, 1.01, 0.01)
precisions = []
recalls = []

# Compute precision and recall for each threshold
for threshold in thresholds:
    y_val_pred = (y_val_pred_proba >= threshold).astype(int)
    precision = precision_score(y_val, y_val_pred, zero_division=0)
    recall = recall_score(y_val, y_val_pred, zero_division=0)
    precisions.append(precision)
    recalls.append(recall)

# Find where precision and recall curves intersect
intersection_threshold = None
min_diff = float('inf')

for i, threshold in enumerate(thresholds):
    diff = abs(precisions[i] - recalls[i])
    if diff < min_diff:
        min_diff = diff
        intersection_threshold = threshold

print(f"Precision and recall curves intersect at threshold: {intersection_threshold:.2f}")

# Plot the precision-recall curves
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions, label='Precision', linewidth=2)
plt.plot(thresholds, recalls, label='Recall', linewidth=2)
plt.axvline(x=intersection_threshold, color='red', linestyle='--', 
            label=f'Intersection at threshold = {intersection_threshold:.2f}')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision and Recall vs. Threshold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Print the precision and recall values at the intersection point
intersection_idx = np.argmin(np.abs(np.array(precisions) - np.array(recalls)))
print(f"At threshold {intersection_threshold:.2f}:")
print(f"  Precision = {precisions[intersection_idx]:.4f}")
print(f"  Recall = {recalls[intersection_idx]:.4f}")
```

```python
# Find the exact intersection point with more precision
intersection_idx = np.argmin(np.abs(np.array(precisions) - np.array(recalls)))
exact_threshold = thresholds[intersection_idx]
exact_precision = precisions[intersection_idx]
exact_recall = recalls[intersection_idx]

print(f"Exact intersection point:")
print(f"  Threshold: {exact_threshold:.3f}")
print(f"  Precision: {exact_precision:.4f}")
print(f"  Recall: {exact_recall:.4f}")
```

```python
from sklearn.metrics import f1_score

# Generate predictions on validation set
y_val_pred_proba = model.predict_proba(X_val_encoded)[:, 1]

# Initialize lists to store F1 scores
thresholds = np.arange(0.0, 1.01, 0.01)
f1_scores = []

# Compute F1 score for each threshold
for threshold in thresholds:
    y_val_pred = (y_val_pred_proba >= threshold).astype(int)
    f1 = f1_score(y_val, y_val_pred, zero_division=0)
    f1_scores.append(f1)

# Find the threshold with maximum F1 score
max_f1_idx = np.argmax(f1_scores)
max_f1_threshold = thresholds[max_f1_idx]
max_f1_value = f1_scores[max_f1_idx]

print(f"Maximum F1 score: {max_f1_value:.4f}")
print(f"Achieved at threshold: {max_f1_threshold:.2f}")

# Let's also get precision and recall at this optimal threshold
y_val_pred_optimal = (y_val_pred_proba >= max_f1_threshold).astype(int)
optimal_precision = precision_score(y_val, y_val_pred_optimal, zero_division=0)
optimal_recall = recall_score(y_val, y_val_pred_optimal, zero_division=0)

print(f"At optimal threshold {max_f1_threshold:.2f}:")
print(f"  Precision: {optimal_precision:.4f}")
print(f"  Recall: {optimal_recall:.4f}")
print(f"  F1 Score: {max_f1_value:.4f}")
```

```python
from sklearn.model_selection import KFold



print(f"Full dataset size: {len(df_full_train)}")

# Convert to dictionary format for DictVectorizer
full_dict = df_full_train.to_dict(orient='records')

# Initialize KFold
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Initialize lists to store AUC scores
auc_scores = []

# Perform 5-fold cross-validation
for fold, (train_idx, val_idx) in enumerate(kf.split(full_dict)):
    print(f"Processing fold {fold + 1}...")
    
    # Split data
    X_train_fold = [full_dict[i] for i in train_idx]
    X_val_fold = [full_dict[i] for i in val_idx]
    y_train_fold = y_full_train[train_idx]
    y_val_fold = y_full_train[val_idx]
    
    # Apply DictVectorizer
    dv = DictVectorizer(sparse=False)
    X_train_encoded = dv.fit_transform(X_train_fold)
    X_val_encoded = dv.transform(X_val_fold)
    
    # Train model
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=1)
    model.fit(X_train_encoded, y_train_fold)
    
    # Predict and calculate AUC
    y_val_pred = model.predict_proba(X_val_encoded)[:, 1]
    auc = roc_auc_score(y_val_fold, y_val_pred)
    auc_scores.append(auc)
    
    print(f"  Fold {fold + 1} AUC: {auc:.4f}")

# Calculate statistics
mean_auc = np.mean(auc_scores)
std_auc = np.std(auc_scores)

print("\nCross-validation results:")
print(f"AUC scores: {[f'{score:.4f}' for score in auc_scores]}")
print(f"Mean AUC: {mean_auc:.4f}")
print(f"Standard deviation: {std_auc:.4f}")
print(f"Range: {min(auc_scores):.4f} - {max(auc_scores):.4f}")
```

```python
uestion 6: Hyperparameter Tuning

Now let's use 5-Fold cross-validation to find the best parameter C

    Iterate over the following C values: [0.000001, 0.001, 1]
    Initialize KFold with the same parameters as previously
    Use these parameters for the model: LogisticRegression(solver='liblinear', C=C, max_iter=1000)
    Compute the mean score as well as the std (round the mean and std to 3 decimal digits)

Which C leads to the best mean score?
```

```python
from sklearn.model_selection import KFold


full_dict = df_full_train.to_dict(orient='records')

# C values to test
C_values = [0.000001, 0.001, 1]
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Store results
results = {}

for C in C_values:
    print(f"Testing C = {C}")
    auc_scores = []
    
    for fold, (train_idx, val_idx) in enumerate(kf.split(full_dict)):
        # Split data
        X_train_fold = [full_dict[i] for i in train_idx]
        X_val_fold = [full_dict[i] for i in val_idx]
        y_train_fold = y_full_train[train_idx]
        y_val_fold = y_full_train[val_idx]
        
        # Apply DictVectorizer
        dv = DictVectorizer(sparse=False)
        X_train_encoded = dv.fit_transform(X_train_fold)
        X_val_encoded = dv.transform(X_val_fold)
        
        # Train model with current C
        model = LogisticRegression(solver='liblinear', C=C, max_iter=1000, random_state=1)
        model.fit(X_train_encoded, y_train_fold)
        
        # Predict and calculate AUC
        y_val_pred = model.predict_proba(X_val_encoded)[:, 1]
        auc = roc_auc_score(y_val_fold, y_val_pred)
        auc_scores.append(auc)
    
    # Calculate statistics
    mean_auc = np.mean(auc_scores)
    std_auc = np.std(auc_scores)
    
    results[C] = {
        'mean': round(mean_auc, 3),
        'std': round(std_auc, 3),
        'scores': auc_scores
    }
    
    print(f"  C = {C}: Mean AUC = {mean_auc:.3f}, Std = {std_auc:.3f}")

# Find the best C
best_C = max(results.keys(), key=lambda x: results[x]['mean'])
best_mean = results[best_C]['mean']
best_std = results[best_C]['std']

print(f"\nBest C: {best_C}")
print(f"Best mean AUC: {best_mean}")
print(f"Standard deviation: {best_std}")

# Display all results
print("\nDetailed results:")
for C in C_values:
    print(f"C = {C}: Mean = {results[C]['mean']}, Std = {results[C]['std']}")
```
