```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('course_lead_scoring.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe(include='object')
```

```python
categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
categorical_columns
```

```python
numerical_columns = list(df.dtypes[df.dtypes != 'object'].index)
```

```python
numerical_columns = ['number_of_courses_viewed', 'annual_income', 'interaction_count', 'lead_score']
```

```python
numerical_columns
```

```python
for n in numerical_columns:
    df[n] = df[n].fillna(0)
```

```python
for c in categorical_columns:
    df[c] = df[c].fillna('NA')
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.converted.values
y_val = df_val.converted.values
y_test = df_test.converted.values

del df_train['converted']
del df_val['converted']
del df_test['converted']
```

```python
# Compute AUC for each numerical variable
auc_scores = {}

for feature in numerical_columns:
    # Calculate AUC
    auc = roc_auc_score(y_train, df_train[feature])
    
    # If AUC < 0.5, invert the variable
    if auc < 0.5:
        auc = roc_auc_score(y_train, -df_train[feature])
        auc_scores[feature] = auc
        print(f"{feature}: {auc:.4f} (inverted)")
    else:
        auc_scores[feature] = auc
        print(f"{feature}: {auc:.4f}")

# Find the numerical variable with the highest AUC
best_feature = max(auc_scores, key=auc_scores.get)
highest_auc = auc_scores[best_feature]

print(f"\nThe numerical variable with the highest AUC is '{best_feature}' with AUC = {highest_auc:.4f}")
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical_columns + numerical_columns].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical_columns + numerical_columns].to_dict(orient='records')
X_val = dv.transform(val_dict)

y_pred = model.predict_proba(X_val)[:, 1]
conversion_decision = (y_pred >= 0.5)
(y_val == conversion_decision).mean()
```

```python
auc = roc_auc_score(y_val, y_pred)

print(round(auc, 3))
```

```python
preprocessor_all = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', ['number_of_courses_viewed', 'annual_income', 'interaction_count', 'lead_score']),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), ['lead_source', 'industry', 'employment_status', 'location'])
    ])

X_train_all = preprocessor_all.fit_transform(df_train)
X_val_all = preprocessor_all.transform(df_val)

model_all = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
model_all.fit(X_train_all, y_train)
y_val_pred_all = model_all.predict(X_val_all)
original_accuracy = accuracy_score(y_val, y_val_pred_all)

print(f"Original accuracy with all features: {original_accuracy:.4f}")

# Test each feature by excluding it
accuracy_differences = {}

for feature_to_exclude in full_column:
    print(f"\nExcluding feature: {feature_to_exclude}")
    
    # Determine which features to keep
    features_to_keep = [f for f in full_column if f != feature_to_exclude]
    
    # Separate categorical and numerical features to keep
    cat_features_to_keep = [f for f in features_to_keep if f in categorical_columns]
    num_features_to_keep = [f for f in features_to_keep if f in numerical_columns]
    
    # Create preprocessor for this feature set
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', 'passthrough', num_features_to_keep),
            ('cat', OneHotEncoder(drop='first', sparse_output=False), cat_features_to_keep)
        ])
    
    # Apply preprocessing
    X_train_processed = preprocessor.fit_transform(df_train)
    X_val_processed = preprocessor.transform(df_val)
    
    # Train model
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model.fit(X_train_processed, y_train)
    
    # Make predictions and calculate accuracy
    y_val_pred = model.predict(X_val_processed)
    accuracy = accuracy_score(y_val, y_val_pred)
    
    # Calculate difference from original accuracy
    difference = original_accuracy - accuracy
    accuracy_differences[feature_to_exclude] = difference
    
    print(f"Accuracy without {feature_to_exclude}: {accuracy:.4f}")
    print(f"Difference from original: {difference:.4f}")

# Find the feature with the smallest difference
smallest_diff_feature = min(accuracy_differences, key=lambda x: abs(accuracy_differences[x]))
smallest_diff_value = accuracy_differences[smallest_diff_feature]

print(f"\n{'='*50}")
print("Summary of accuracy differences:")
for feature, diff in sorted(accuracy_differences.items(), key=lambda x: abs(x[1])):
    print(f"{feature}: {diff:.4f}")

print(f"\nThe feature with the smallest difference is: '{smallest_diff_feature}'")
print(f"Difference: {smallest_diff_value:.4f}")
```
